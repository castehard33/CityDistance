# **Projekt MSI**

# **Loading dataset**

```python
import pandas as pd
from IPython.display import display

df = pd.read_csv('https://raw.githubusercontent.com/castehard33/CityDistance/main/Drugi_Projekt/delaney_solubility_with_descriptors.csv', sep=',')

print("DataFrame wczytany.")
display(df)
```

# **Data preparation**

## Data separation as X and y

```python
y = df['logS']
y
```

```python
X = df.drop('logS', axis=1)
X
```

## Data splitting

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

# **Model Building**

## **Linear Regression**

### **Training the model**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

### **Applying the model to make a prediction**

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### **Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score   #błąd średniokwadratowy i współczynnik determinacji

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

### **Linear Regression with Cross Validation**

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import numpy as np

lr_cv = LinearRegression()
# cv=10 oznacza 10-krotną walidację krzyżową
# scoring='r2' - chcemy uzyskać wyniki dla metryki R2
# scoring='neg_mean_squared_error' - dla MSE (negatywny, bo cross_val_score maksymalizuje)


lr_r2_scores = cross_val_score(lr_cv, X, y, cv=10, scoring='r2', n_jobs=-1)
lr_mse_scores = cross_val_score(lr_cv, X, y, cv=10, scoring='neg_mean_squared_error', n_jobs=-1)


mean_r2_lr = np.mean(lr_r2_scores)
std_r2_lr = np.std(lr_r2_scores)
mean_mse_lr = -np.mean(lr_mse_scores)
std_mse_lr = np.std(lr_mse_scores)


cv_summary_lr_df = pd.DataFrame(
    data={
        'Mean': [np.mean(lr_r2_scores), -np.mean(lr_mse_scores)],
        'Std Dev': [np.std(lr_r2_scores), np.std(lr_mse_scores)]
    },
    index=['R2 Score (LR)', 'MSE (LR)']
)


cv_summary_lr_df

```

## **Random Forest**

### **Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=5, random_state=100)
rf.fit(X_train, y_train)
```

### **Applying the model to make a prediction**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

### **Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

## **Model comparison**

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
```

```python
df_models.reset_index(drop=True)
```

# **Data visualization of prediction results**

```python
import matplotlib.pyplot as plt
import numpy as np

# LR
plt.figure(figsize=(6, 5))

plt.scatter(x=y_train, y=y_lr_train_pred, c="#0d089c", alpha=0.5, label='LR Predictions')

z_lr = np.polyfit(y_train, y_lr_train_pred, 1)
p_lr = np.poly1d(z_lr)
x_trend_values_lr = np.array([min(y_train), max(y_train)]) # Użycie min i max y_train
plt.plot(x_trend_values_lr, p_lr(x_trend_values_lr), color='#FF0000', linestyle='-', linewidth=2, label='LR Trend Line')

plt.title('Linear Regression: Actual vs. Predicted (Training Data)')
plt.xlabel('Experimental LogS (Actual)')
plt.ylabel('Predict LogS (Predicted by LR)')
plt.grid(True)
plt.legend()
plt.show()

print("\n" * 3)

# RF
plt.figure(figsize=(6, 5))

plt.scatter(x=y_train, y=y_rf_train_pred, c="green", marker='o', alpha=0.5, label='RF Predictions')

z_rf = np.polyfit(y_train, y_rf_train_pred, 1)
p_rf = np.poly1d(z_rf)
x_trend_values_rf = np.array([min(y_train), max(y_train)]) # Użycie min i max y_train
plt.plot(x_trend_values_rf, p_rf(x_trend_values_rf), color='darkgreen', linestyle='-', linewidth=2, label='RF Trend Line')

plt.title('Random Forest: Actual vs. Predicted (Training Data)')
plt.xlabel('Experimental LogS (Actual)')
plt.ylabel('Predict LogS (Predicted by RF)')
plt.grid(True)
plt.legend()
plt.show()
```
